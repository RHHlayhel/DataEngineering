```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots 
```

```python
df = spark.table("silk_prices").toPandas()
df = df.sort_values("timestamp")

print(f"Total data points: {len(df)}")
df.head()
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Silk Price Over Time',
        'Price Distribution',
        'PPI Index Over Time',
        'PPI vs Silk Price'
    )
)

# 1. Price trend over time
fig.add_trace(go.Scatter(
    x    = df['timestamp'],
    y    = df['price_usd_mt'],
    mode = 'lines+markers',
    name = 'Silk Price (USD/MT)',
    line = dict(color='#2ecc71', width=2),
    marker = dict(size=8)
), row=1, col=1)

# 2. Price distribution
fig.add_trace(go.Histogram(
    x      = df['price_usd_mt'],
    nbinsx = 10,
    name   = 'Price Distribution',
    marker_color = 'steelblue',
    showlegend   = False
), row=1, col=2)

# 3. PPI index over time
fig.add_trace(go.Scatter(
    x    = df['ppi_date'],
    y    = df['ppi_value'],
    mode = 'lines+markers',
    name = 'PPI Index',
    line = dict(color='#e67e22', width=2),
    marker = dict(size=8)
), row=2, col=1)

# 4. PPI vs Silk Price scatter
fig.add_trace(go.Scatter(
    x    = df['ppi_value'],
    y    = df['price_usd_mt'],
    mode = 'markers',
    name = 'PPI vs Price',
    marker = dict(
        size  = 10,
        color = '#9b59b6'
    ),
    showlegend = False
), row=2, col=2)

fig.update_layout(
    title_text      = 'Silk Price Monitoring Dashboard',
    title_font_size = 20,
    height          = 700,
    template        = 'plotly_white'
)

fig.show()
```
